```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>5
e_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(e_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=1
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
def recommend(user_id):
    if user_id in pt.columns:
        
        user_index = np.where(pt.columns == user_id)[0][0]
        similar_users = sorted(list(enumerate(similarity_score[user_index])), key=lambda x: x[1], reverse=True)[1:6]
        
        data = []
        for i in similar_users:
            similar_user_id = pt.columns[i[0]]
            user_books = ratings[ratings['User-ID'] == similar_user_id]
            top_books = user_books[user_books['Book-Rating'] >= 4]
            
            for _, book in top_books.iterrows():
                item = []
                temp_df = books[books['Book-Title'] == book['Book-Title']]
                item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
                item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
                item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
                data.append(item)
        
        return data
    
    else:
        
        print(f"New user detected. Recommending popular books.")
        popular_books = books['Book-Title'].value_counts().index[:5]
        data = []
        for book in popular_books:
            item = []
            temp_df = books[books['Book-Title'] == book]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
            data.append(item)
        
        return data
```
